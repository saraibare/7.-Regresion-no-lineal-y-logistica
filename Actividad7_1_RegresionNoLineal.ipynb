```python
#Cargamos librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.special as special
from scipy.optimize import curve_fit
import seaborn as sns
from sklearn.metrics import r2_score
```

```python
#Cargar archivo csv
gastos2020=pd.read_csv("gastos2020_iqr.csv")
gastos2021=pd.read_csv("gastos2021_iqr.csv")
gastos2022=pd.read_csv("gastos2022_iqr.csv")
gastos2023=pd.read_csv("gastos2023_iqr.csv")
```

## **Gastos y costos 2020**

```python
#Graficamos todas las dispersiones entre todas las variables
sns.pairplot(gastos2020)
```

```python
#Declaramos las variables dependientes e independientes para la regresi√≥n No lineal
x= gastos2020[['IVA']]
y= gastos2020['TOTAL MX']
```

### **MODELO 1: y = ax^2 + bx + c**

```python
def  func1 (x, a, b, c):
     return a*x**2 + b*x + c
```

```python
#Ajustamos los par√°metros de la funci√≥n curve_fit
parametros, covs= curve_fit(func1, gastos2020['IVA'], gastos2020['TOTAL MX'])
```

```python
#Obtenemos los coeficientes del modelo de regresi√≥n no lineal
parametros
```

```python
#Creamos el modelo de predicci√≥n con los par√°metros obtenidos 
parametros, _ = curve_fit(func1, gastos2020['IVA'], gastos2020['TOTAL MX'])
a, b, c = parametros[ 0 ], parametros[ 1 ], parametros[ 2 ]
yfit1 = a * x ** 2 + b * x + c
```

```python
#Calculamos las predicciones y reestructuramos el vector de predicciones
#x= x.reshape(51, 1)
#y = y.reshape(51, 1)
#yfit1 = yfit1.reshape(51, 1)
yfit1
```

```python
#Graficamos las predicciones y los datos originales para realizar la comparaci√≥n
plt.plot(x, y, 'bo', label="y-original")
plt.plot(x, yfit1, label="y=a*x^2+b*x+c")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best', fancybox=True, shadow=True)
plt.grid(True)
plt.show() 
```

```python
#Calculamos el coeficiente de determinaci√≥n del modelo 1
R2_2020_1 = r2_score(y, yfit1) 
R2_2020_1
```

De acuerdo a la escala de correlaci√≥n, el coeficiente obtenido se encuentra entre una correlaci√≥n positiva considerable y una correlaci√≥n positiva muy fuerte, por lo que es un buen modelo para predecir la variable IVA respecto a la variable TOTAL MX

```python
#Corroboramos cual es el coeficiente de Correlaci√≥n del modelo 1
Correl_2020_1=np.sqrt(R2_2020_1)
Correl_2020_1
```

### **MODELO 2: y = ax^3 + bx^2 + c**

```python
def  func2 (x, a, b, c):
     return a*x**3 + b*x**2 + c
```

```python
#Ajustamos los par√°metros de la funci√≥n curve_fit
parametros2, covs2= curve_fit(func2, gastos2020['IVA'], gastos2020['TOTAL MX'])
```

```python
#Creamos el modelo de predicci√≥n con los par√°metros obtenidos 
parametros2, _ = curve_fit(func2, gastos2020['IVA'], gastos2020['TOTAL MX'])
a2, b2, c2 = parametros2[ 0 ], parametros2[ 1 ], parametros2[ 2 ]
yfit2 = a2 * x ** 2 + b2 * x ** 2 + c2
```
